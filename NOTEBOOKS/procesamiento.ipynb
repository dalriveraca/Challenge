```python
import yaml
import pyodbc
import pandas as pd
from jinja2 import Template
import numpy as np
import os
```

```python

path = "C:/Users/diegrica/OneDrive - colsubsidio.com/Documents/DARC/MELI/NOTEBOOKS/"
ruta_yml = os.path.join(path, "config.yml")
with open(ruta_yml, "r") as f:
    config = yaml.safe_load(f)

params = config["parametros"]







```

Leer el archivo .sql como código Python y extraer el string

```python

namespace = {}
ruta_sql = os.path.join(path, "query_challenge.sql")
with open(ruta_sql, "r", encoding="utf-8") as f:
    exec(f.read(), namespace)
```

Cargar el query

```python

query_prints_str = namespace["query_prints"]
```

Renderizar el query .sql

```python

query_prints_ren = Template(query_prints_str).render(**params)
```

Conexión a SQL Server

```python

with pyodbc.connect(
    r'DRIVER={ODBC Driver 17 for SQL Server};'
    r'SERVER=BOGAL06DIEGRICA;'
    r'DATABASE=Challenge;'
    r'Trusted_Connection=yes;'
) as conn:
    prints = pd.read_sql(query_prints_ren, conn)
```
